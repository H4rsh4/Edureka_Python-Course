# Type of Joins:
**Inner:**
    - -
**Outer:**
    - Left
    - Right
    - Full

```python
inner_joined_df = MATH.join(PHYS, how='inner', lsuffix='_PHYS',)
inner_joined_df
```

```python
left_outer_joined_df = MATH.join(PHYS, how='left', rsuffix='_Phys',)
left_outer_joined_df
```

```python
right_outer_joined_df = MATH.join(PHYS, how='right', rsuffix='_Phys')
right_outer_joined_df
```

```python
full_outer_joined_df = MATH.join(PHYS, how='outer', rsuffix='_Phys')
full_outer_joined_df
```
