```python
#Part 2 : Data Manipulation
```

```python
totalBenefits = []

for state in range(len(df_raw['STATE/UT'])):
    totalBenefits.append(
        df_raw.loc[state][2:].sum()
    )
totalBenefits
```

```python
df_raw['total_benefitted'] = totalBenefits
```

```python
df_raw.transpose()
```

```python
stuff_to_not_add = ['STATE/UT', 'YEAR']
df_raw.append(
    pd.Series(
        {
            col:(df_raw[col].sum() if col not in stuff_to_not_add else np.nan)\
            for col in df_raw.columns
        },
        name='totals'
    ),
)
```

```python
#Plotting:
```

```python
fig,axe = plt.subplots(figsize=(20,10), dpi=50)
#bar plot divided into two parts
axe.bar(df_raw['STATE/UT'][:int(len(df_raw)/2)], df_raw['total_benefitted'][:int(len(df_raw)/2)])
#plt.autoscale(True)
```

```python
fig,axe = plt.subplots(figsize=(20,10), dpi=50)
axe.bar(df_raw['STATE/UT'][int(len(df_raw)/2):], df_raw['total_benefitted'][int(len(df_raw)/2):])
```
