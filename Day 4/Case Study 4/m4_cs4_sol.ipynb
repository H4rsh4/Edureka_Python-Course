```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_PATH = 'E:/Edureka_Python-Course/Data Sources/Day_4/'
```

```python
df_raw = pd.read_csv(DATA_PATH+'BigMartSalesData.csv', low_memory=False)
df_raw
```

```python
#Total Sales Per Month for Year 2011
data_from_2011 = df_raw[df_raw['Year'] == 2011]
data_from_2011
```

```python
x_q1 = [round(sum(data_from_2011.Amount[data_from_2011.Month == i]), 2) for i in sorted(data_from_2011.Month.unique())]
y_q1 = sorted(data_from_2011.Month.unique())
```

```python
x_q1
```

```python
plt.plot(y_q1, x_q1, 'b-')

#Month 2 has the lowest Sales
```

```python
months_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep' ,'Oct', 'Nov', 'Dec' ]
bar_plot = plt.bar(y_q1, x_q1, tick_label=months_names, )

#Both are good various puposed; Bar chart is good for visualizing sales per month
#And the basic plot can help identify trends in the data much better
```

```python
labels = data_from_2011.Country.unique()
sizes = [round(sum(data_from_2011.Amount[data_from_2011.Country == i]), 2) for i in labels]
explode = [0.1 if i != max(sizes) else 0 for i in sizes]

plt.pie(sizes, explode=explode, labels=labels, startangle=90, shadow=True)

#United Kingdom is highest contributor
```
